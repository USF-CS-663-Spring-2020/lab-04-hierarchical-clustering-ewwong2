# Lab 4 - Dendrogram

## US Census Data

```python
import pandas as pd

# The data is wrapped with this notebook.
# This is a smaller version of the 1990 census, available at UCI and other places.
# Taking a subset helps running time and to see the dendrogram.
df = pd.read_csv('USCensus1990-2000rows.data.txt')
df
```

```python
# While the implementation in sklearn is good,
# the scipy version can produce the dendrogram
from scipy.cluster.hierarchy import linkage, dendrogram

Z = linkage(df, 'ward')
```

```python
# Draw the dendrogram
import matplotlib.pyplot as plt

%matplotlib inline

plt.figure()
dendrogram(Z)
plt.show()
```

## Nips Dataset

```python
import pandas as pd
nips_df = pd.read_csv('NIPS_1987-2015.csv')
nips_df = nips_df.iloc[:, 1:]
nips_df
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram

nips_Z = linkage(nips_df, 'ward')

# Draw the dendrogram
import matplotlib.pyplot as plt

%matplotlib inline

plt.figure()
dendrogram(nips_Z)
plt.show()
```
